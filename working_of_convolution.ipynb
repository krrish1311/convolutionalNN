```python
# library to get free images 
from scipy import misc
img = misc.ascent()
```

```python
img.shape
```

```python
img
```

```python
plt.plot(img)
```

```python
plt.imshow(img, cmap='gray')
```

```python
# creating filter
# Convolution layer in CNN will automatically create filter. 
fltr = np.array([
    [1,2,3],
    [0,2,0],
    [-2,-4,-5]
])
```

```python
fltr.shape
```

```python
# convolve2d to convolve 2D array
from scipy.signal import convolve2d
conv_img=convolve2d(img,fltr)
```

```python
# Convolved Image 
conv_img
```

```python
plt.plot(conv_img)
```

```python
plt.imshow(conv_img, cmap='gray')
```

```python
conv_img.shape
```
